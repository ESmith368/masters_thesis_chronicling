```python
from bs4 import BeautifulSoup as bs
import pandas as pd
from lxml import etree
```

```python
'''reading in the file'''

content = []
# Read the XML file
with open("C:/Users/Ellie/Documents/MASTERS/NIAA/new_allie/1791_Purm_Louw_kopie_tei.xml", "r", encoding= 'utf-8') as file:
    # Read each line in the file, readlines() returns a list of lines
    content = file.readlines()
    # Combine the lines in the list into a string
    content = "".join(content)
    bs_content = bs(content, "lxml")
```

```python
'''method 1: find_all on kopie tag'''

result = bs_content.find_all("kopie")

line_content_K =list()
for item in result:
    content= item.text
    line_content_K.append(content)
    
print(len(line_content_K))
```

```python
'''method 2: find_all l and comparing the content of method 1. Shows that there are duplicates where some are not tagged as 
kopie in method 1'''

all_l = bs_content.find_all('l')
content_l =list()
for item in all_l:
    content= item.text
    content_l.append(content)

kopies_in_l =list()
for kopie in line_content_K:
    for line in content_l:
        if kopie == line:
            kopies_in_l.append(kopie)
            
print(len(kopies_in_l))
```

```python
'''method 3: find_all l then convert to string and search for the kopie tag as a string'''

dict_list = list()
n = 0
everything_results = bs_content.find_all("l")
for item in list(everything_results):
    item_string = str(item)
    temp_dict =dict()
    n+=1
    if '<kopie>' in item_string:
        temp_dict['text'] = item.text
        temp_dict['index']= n
        temp_dict['kopie']= 1
        dict_list.append(temp_dict)
    else:
        temp_dict['text'] =item.text
        temp_dict['index'] = n
        temp_dict['kopie'] = 0
        dict_list.append(temp_dict)
        
'''checking kopie number'''
method_3 = list()
for item in dict_list:
    if item['kopie'] == 1:
        method_3.append(item['text'])
        
print(len(method_3))
```

```python
'''method 4 using etree '''

tree = etree.parse("C:/Users/Ellie/Documents/MASTERS/NIAA/new_allie/1791_Purm_Louw_kopie_tei.xml")
root = tree.getroot()
body = root[1]

kopie_text = list()
for elem in body: 
    for e in elem.iter():
        if 'kopie' in str(e):
            kopie_text.append(e.text)
        
print(len(kopie_text))
```

```python
'''renaming outputs for clarity'''
method_1_list = line_content_K
method_2_list = kopies_in_l
method_3_list = method_3
method_4_list = kopie_text
```

```python
found_kopies_set =  set(method_1_list+method_2_list+method_3_list)

main_list = list()
for line in found_kopies_set:
    line_dict =dict()
    line_dict['line'] = line
    
    if line in method_1_list:
        line_dict['method_1'] = 1
    else:
        line_dict['method_1'] = 0
        
    if line in method_2_list:
        line_dict['method_2'] = 1
    else:
        line_dict['method_2'] = 0
        
    if line in method_3_list:
        line_dict['method_3'] = 1
    else:
        line_dict['method_3'] = 0
        
    if line in method_4_list:
        line_dict['method_4'] = 1
    else:
        line_dict['method_4'] = 0
        
    main_list.append(line_dict)
    
df = pd.DataFrame(main_list)
```
