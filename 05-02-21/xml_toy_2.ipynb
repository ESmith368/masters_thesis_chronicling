```python
from bs4 import BeautifulSoup as bs

content = []
# Read the XML file
with open("1791_Purm_Louw_kopie_tei.xml", "r", encoding= 'utf-8') as file:
    # Read each line in the file, readlines() returns a list of lines
    content = file.readlines()
    # Combine the lines in the list into a string
    content = "".join(content)
    bs_content = bs(content, "lxml")
```

```python
everything_results = bs_content.find_all("l")
```

```python
dict_list = list()
n = 0
for item in list(everything_results):
    item_string = str(item)
    temp_dict =dict()
    n+=1
    if '<kopie>' in item_string:
        temp_dict['text'] = item.text
        temp_dict['index']= n
        temp_dict['kopie']= 1
        dict_list.append(temp_dict)
    else:
        temp_dict['text'] =item.text
        temp_dict['index'] = n
        temp_dict['kopie'] = 0
        dict_list.append(temp_dict)
#print(len(dict_list))
#print(dict_list[0:10])

'''I like this list of dicts and it gives me everything I need, but Im still not sure if the kopies line up between this and 
the find_all(kopie) method. In theory they should because everythings is first tagged as a linebefore a nested kopie tag
...The difference in length between the two is 27'''

dict_kopie_list = list()
for temp_dict in dict_list:
    if temp_dict['kopie'] == 1:
        dict_kopie_list.append(temp_dict['text'])
        
#print(len(dict_kopie_list))
#print(len(line_content_K))

#print(len(set(dict_kopie_list)-(set(line_content_K))))
'''there are 119 kopies which are not shared between the texts. This is different from their difference in length being 27...'''
kopie_indices = list()
for temp_dict in dict_list:
    if temp_dict['kopie'] == 1:
        kopie_indices.append(temp_dict['index'])
#print(kopie_indices)
'''unlike in the original attempt, here the indices line up perfectly e.g. consecutive lines which are copied have
consecutive indices.'''
```

```python
'''Double checking the number of kopies vs non-kopies in new_df'''
kopie_freqs=new_df['kopie'].value_counts()
kopie_freqs_df = pd.DataFrame(kopie_freqs.iteritems(), columns =['Kopie', 'Freq',])
kopie_freqs_df
```

```python
'''working on grouping from new_df'''

'''seperating kopies and non-kopies into seperate dfs'''
kopie_bool = new_df['kopie'] == 1
kopie_df = new_df[kopie_bool]
#print(kopie_df[0:5])
no_kopie_bool = new_df['kopie'] == 0
no_kopie_df = new_df[no_kopie_bool]
#print(no_kopie_df[0:5])

'''finding groups of consecutive indices in each seperated df'''
def ranges(nums):
    nums = sorted(set(nums))
    gaps = [[s, e] for s, e in zip(nums, nums[1:]) if s+1 < e]
    edges = iter(nums[:1] + sum(gaps, []) + nums[-1:])
    return list(zip(edges, edges))

index_groups_kopie = ranges(kopie_df['index'])
index_groups_no_kopie = ranges(no_kopie_df['index'])
```

```python
'''Using the index group ranges (list of tuples) to find the rows from the corresponding indices in new_df'''
#kopie_main_list = list()
#for group in index_groups_kopie:
#    grouping_dict = dict()
#    text_list = list()
#    for index, row in new_df.iterrows():
#        for i in range (group[0],group[1]+1):
#            if row['index'] == i:
#                text_list.append(row['text'])
#    grouping_dict['text_list'] = text_list
#    grouping_dict['index_range'] = group
#    grouping_dict['kopie'] = 1
#    kopie_main_list.append(grouping_dict)
    
#print(kopie_main_list[0:2])
'''the above method is much too slow to give results'''
#kopie_main_list =list()
#for group in index_groups_kopie:
#    grouping_dict =dict()
#    grouping_dict['index_range'] = group
#    grouping_dict['kopie'] = 1
#    text_list =list()
#    for index, row in new_df.iterrows():
#        if row['index'] in range (group[0], group[1]+1):
#            text_list.append(row['text'])
#    grouping_dict['text_list'] = text_list
#    kopie_main_list.append(grouping_dict)
            
#print(kopie_main_list[0:2])
''' the above method is also too slow to produce results. Try to break this in two and use the .loc or .index pandas function
to find each of the rows first before taking the infomration from them.Like in dig_hum.'''
loc_main_K = list()
for group in index_groups_kopie:
    loc_group_dict =dict()
    loc_group_dict['index_range'] = group
    loc_list = list()
    for i in range (group[0],group[1]+1):
        loc_list += list(new_df.loc[new_df['index'] == i].index)
    loc_group_dict['loc_list'] = loc_list
    loc_main_K.append(loc_group_dict)
    
#print(loc_main_K[0:2])

loc_main_NK = list()
for group in index_groups_no_kopie:
    loc_group_dict =dict()
    loc_group_dict['index_range'] = group
    loc_list = list()
    for i in range (group[0],group[1]+1):
        loc_list += list(new_df.loc[new_df['index'] == i].index)
    loc_group_dict['loc_list'] = loc_list
    loc_main_NK.append(loc_group_dict)
    
#print(loc_main_NK[0:2])
```

```python
'''Now we have two list of dicts, one for kopie and one for non-kopie. For each the dicts consist of index_range which is the
line index range and should be preserved an included and loc_list which is the list of row locations which the text needs to 
be collected from in new_df for that chunk.'''
K_list = list()
for group_dict in loc_main_K:
    temp_dict = dict()
    group_text = list(new_df.loc[group_dict['loc_list'], 'text'])
    temp_dict['index_range'] = group_dict['index_range']
    temp_dict['kopie'] = 1
    temp_dict['text_list'] = "".join(group_text)
    K_list.append(temp_dict)
    
#print(K_list[0:3])

NK_list = list()
for group_dict in loc_main_NK:
    temp_dict = dict()
    group_text = list(new_df.loc[group_dict['loc_list'], 'text'])
    temp_dict['index_range'] = group_dict['index_range']
    temp_dict['kopie'] = 0
    temp_dict['text_list'] = "".join(group_text)
    NK_list.append(temp_dict)
    
#print(NK_list[0:3])
```

```python
'''combining the two lists of dicts'''
whole_list= K_list+NK_list
#print(type(whole[0]))
print(whole_list[0])
```
