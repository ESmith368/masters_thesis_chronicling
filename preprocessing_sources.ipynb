```python
import os
import pandas as pd
import re
from pathlib import Path
```

```python
def read_files(directory):
    
    all_files = os.listdir(directory)
    print(f'reading {len(all_files)} files...')
    main = list()
    for file in all_files:
        temp_dict = dict()
        temp_dict['file'] = file
        filepath = directory + '/' + file
        with open(filepath, encoding = 'utf-8') as f:
            lines = f.readlines()
            joined = ' '.join(lines)
            temp_dict['text'] = joined
        main.append(temp_dict)
    df = pd.DataFrame(main) 
    print(f'DataFrame with shape {df.shape} now ready')
    print()
    print(df[0:2])
    
    return df
```

```python
df= read_files("C:/Users/Ellie/Documents/MASTERS/NIAA/new_allie/05-02-21/10-03-21/source_texts/historical")
```

```python
def clean_texts(df):
    
    df['text'] = df['text'].str.replace('\d+', '')
    df['text'] = df['text'].str.replace('\n', '')
    for i, row in df.iterrows():
        text = row['text']
        no_punc = re.sub(r'[^\w\s]','',text)
        df.at[i,'text'] = no_punc
    
    print('texts_cleaned')
    print()
    print(df[0:2])
    
    return df
```

```python
clean_texts(df)
```

```python
def write_results(directory, df):
    
    for index, row in df.iterrows():
        filepath = directory + '/' + 'cleaned_files' 
        Path(filepath).mkdir(parents=True, exist_ok=True)
        opening = open( filepath + '/'+ row['file'], 'w', encoding= 'utf-8')
        opening.writelines(row['text'])
        opening.close()
            
    print(f'{df.shape[0]} cleaned files ready')
```

```python
write_results("C:/Users/Ellie/Documents/MASTERS/NIAA/new_allie/05-02-21/10-03-21/source_texts/historical", df)
```

```python
def main_preprocess(directory):
    
    df= read_files(directory)
    df = clean_texts(df)
    write_results(directory, df)
    
    return df
```
